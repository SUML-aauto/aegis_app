```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
raw_dataset = pd.read_csv("insurance_update.csv", sep=";", usecols=[
    'Make',
    'Model',
    # 'Manufacturing Year',
    'Fuel Type',
    'Engine Power (HP)',
    'Mileage (km)',
    'Number of Accidents',
    'Market Value ($)',
    'Total Owners',
    'Has Dashcam',
    'Vehicles in Family',
    'Driving Experience',
    'AGE',
    'HOMEKIDS',
    'EMPLOYED',
    'INCOME',
    'MSTATUS',
    'GENDER',
    'CAR_TYPE',
    'OLDCLAIM',
    'CLM_FREQ',
    'REVOKED',
    'CLM_AMT',
    'CAR_AGE',
    'CLAIM_FLAG'
])
```

```python
raw_dataset
```

```python
import sklearn.model_selection

train, test = sklearn.model_selection.train_test_split(raw_dataset)
```

```python
train
```

```python
test
```

```python
from sklearn.feature_extraction import FeatureHasher

hasher = FeatureHasher(input_type='string')

def fixup_monetary_string(df, field):
    df[field] = df[field].replace({'\\$': '', ',': ''}, regex=True).astype('float')

def preprocess_data(data):
    data_filtered = data[[
        'Make',
        'Fuel Type',
        # 'Model',
        'Engine Power (HP)',
        'Driving Experience',
        'INCOME',
        'GENDER',
        'CAR_TYPE',
        'CLM_AMT'
    ]].copy()
    data_filtered.fillna(0, inplace=True)

    fixup_monetary_string(data_filtered, ['CLM_AMT', 'INCOME'])

    return pd.get_dummies(data_filtered, columns=[
        'Make', 
        'Fuel Type', 
        # 'Model', 
        'GENDER',
        'CAR_TYPE'
    ])

train_data_preprocessed = preprocess_data(train)
test_data_preprocessed = preprocess_data(test)
```

```python
train_data_preprocessed
```

```python
test_data_preprocessed
```

```python
import sklearn.cluster
import sklearn.ensemble
import sklearn.linear_model

x_train = train_data_preprocessed.loc[:, train_data_preprocessed.columns != 'CLM_AMT']
y_train = train_data_preprocessed['CLM_AMT']

x_test = test_data_preprocessed.loc[:, test_data_preprocessed.columns != 'CLM_AMT']
y_test = test_data_preprocessed['CLM_AMT']

reg = sklearn.linear_model.LinearRegression()
reg = reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

```python
y_pred = reg.predict(x_test)
```
