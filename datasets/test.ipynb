```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import skops.io as sio

pd.set_option('display.max_columns', None)
```

```python
raw_dataset = pd.read_csv("insurance_update.csv", sep=";", usecols=[
    'Make',
    'Model',
    # 'Manufacturing Year',
    'Fuel Type',
    'Engine Power (HP)',
    'Mileage (km)',
    'Number of Accidents',
    'Market Value ($)',
    'Total Owners',
    'Has Dashcam',
    'Vehicles in Family',
    'Driving Experience',
    'AGE',
    'HOMEKIDS',
    'EMPLOYED',
    'INCOME',
    'MSTATUS',
    'GENDER',
    'CAR_TYPE',
    'OLDCLAIM',
    'CLM_FREQ',
    'REVOKED',
    'CLM_AMT',
    'CAR_AGE',
    'CLAIM_FLAG'
])
```

```python
raw_dataset
```

```python
import sklearn.model_selection

train, test = sklearn.model_selection.train_test_split(raw_dataset)
```

```python
train
```

```python
test
```

```python
from sklearn.feature_extraction import FeatureHasher

hasher = FeatureHasher(input_type='string')

def fixup_monetary_string(df, field):
    df[field] = df[field].replace({'\\$': '', ',': ''}, regex=True).astype('float')

def preprocess_data(data):
    data_filtered = data[[
        'Make',
        'Fuel Type',
        # 'Model',
        'Engine Power (HP)',
        'Driving Experience',
        'INCOME',
        'GENDER',
        'CAR_TYPE',
        'CLM_AMT'
    ]].copy()
    data_filtered.fillna(0, inplace=True)

    fixup_monetary_string(data_filtered, ['CLM_AMT', 'INCOME'])

    #data_filtered = pd.get_dummies(data_filtered, columns=[
    #    'Make', 
    #    'Fuel Type', 
    #    # 'Model', 
    #    'GENDER',
    #    'CAR_TYPE'
    #])
    
    data_filtered = data_filtered.assign(class_id = (data_filtered['CLM_AMT'] > 0).astype(int))

    data_filtered = data_filtered.drop('CLM_AMT', axis=1)

    return data_filtered

train_data_preprocessed = preprocess_data(train)
test_data_preprocessed = preprocess_data(test)
```

```python
train_data_preprocessed
```

```python
test_data_preprocessed
```

```python
import sklearn.cluster
from sklearn.compose import make_column_transformer
import sklearn.ensemble
import sklearn.linear_model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

x_train = train_data_preprocessed.loc[:, train_data_preprocessed.columns != 'class_id']
y_train = train_data_preprocessed['class_id']

x_test = test_data_preprocessed.loc[:, test_data_preprocessed.columns != 'class_id']
y_test = test_data_preprocessed['class_id']

one_hot = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), ['Make', 'Fuel Type', 'GENDER', 'CAR_TYPE']),
    remainder='passthrough'
)
model = sklearn.ensemble.GradientBoostingClassifier()
pipeline = Pipeline([
    ('col_transform', one_hot),
    ('model', model)
])
pipeline = pipeline.fit(x_train, y_train)
```

```python
pipeline.score(x_test, y_test)
```

```python
sio.dump(pipeline, 'model.skops')
```
